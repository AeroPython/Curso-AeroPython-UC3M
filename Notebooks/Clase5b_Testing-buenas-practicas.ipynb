## Guía de estilo PEP8

Python ya popularizó la idea de utilizar sangrado para delimitar los bloques de manera estricta. En la comunidad se presta mucha atención a la legibilidad del código, y por eso se desarrolló una guía de estilo llamada PEP8 (*Python Enhancement Proposal 8*) que prácticamente todo el mundo sigue.

http://python.org/dev/peps/pep-0008/

Algunas de estas reglas son:

* Usar sangrado de 4 espacios, no tabuladores [IPython o tu editor se encargan de ello].
* Acotar las líneas a 79 caracteres.
* Usar líneas en blanco para separar funciones y bloques de código dentro de ellas.
* Poner los comentarios en líneas aparte si es posible.
* Usar cadenas de documentación (docstrings).
* Poner espacios alrededor de los operadores y después de coma.
* Usar la convención minuscula_con_guiones_bajos para los nombres de las funciones y las variables.
* Aunque Python 3 te lo permite, no usar caracteres especiales para los identificadores.

Para comprobar si un archivo de código cumple el PEP8 lo mejor es usar el módulo pep8 https://pypi.python.org/pypi/pep8 desde la línea de comandos.

## Profiling

En cuanto empecemos a escribir código numérico de cierta magnitud nuestro programa empezará a ir lento. Puede ser que lleguemos a unos tiempos de ejecución inaceptables (8 horas para ejecutar un programa de un trabajo de clase por ejemplo). ¿Cómo hacer que vaya más rápido?

<blockquote style="font-style: italic; font-size: 1.2em; "><p><sub style="font-size: 3em; ">“</sub> Premature optimization is the root of all evil”.</p>

<p><cite>Donald E. Knuth</cite></p></blockquote>

Existen unas herramientas llamadas *profilers*  que ejecutan nuestro programa y analizan cuánto tiempo tarda cada función dentro de nuestro programa. Podemos usar esta valiosa información para escoger qué partes de nuestro programa tenemos que optimizar *y qué partes **no** tenemos que optimizar*.

```python
import numpy as np
from scipy import optimize

%matplotlib inline
import matplotlib.pyplot as plt
```

```python
%%prun -s cumtime
def kepler(E, e, M):
    """Kepler's equation."""
    return E - e * np.sin(E) - M

# Número de puntos - ¡importante!
N = 100

fig, axes = plt.subplots(figsize=(6, 6))

M_domain = np.linspace(0, 2 * np.pi, N)
for ecc in 0.0167, 0.249, 0.432, 0.775, 0.967:
    E_domain = np.zeros_like(M_domain)
    ii = 0
    for M in M_domain:
        E_domain[ii] = optimize.newton(kepler, E_domain[ii], args=(ecc, M))
        ii += 1

    axes.plot(M_domain, E_domain)

axes.set_xlim(0, 2 * np.pi)
axes.set_ylim(0, 2 * np.pi)
axes.set_xlabel("$M$", fontsize=15)
axes.set_ylabel("$E$", fontsize=15)
axes.set_aspect(1)
axes.grid(True)
axes.legend(["Earth", "Pluto", "Comet Holmes", "28P/Neujmin", "Halley's Comet"], loc=2)
axes.set_title("Kepler's equation solutions")
```
